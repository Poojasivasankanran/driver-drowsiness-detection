## Drowsiness Detection Functions

```python
def calculate_ear(eye):
    A = dist.euclidean(eye[1], eye[5])
    B = dist.euclidean(eye[2], eye[4])
    C = dist.euclidean(eye[0], eye[3])
    return (A + B) / (2.0 * C)

def calculate_mar(mouth):
    A = dist.euclidean(mouth[3], mouth[7])
    B = dist.euclidean(mouth[2], mouth[6])
    C = dist.euclidean(mouth[0], mouth[4])
    return (A + B) / (2.0 * C)
```

## Brightness Adjustment and Logging

```python
def adjust_brightness(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    value = np.mean(hsv[:, :, 2])
    correction_factor = 130 / (value + 1e-5)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * correction_factor, 0, 255)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

def log_event(message):
    with open(log_file, "a") as file:
        file.write(f"[{datetime.datetime.now()}] {message}\n")
```

## Final code

```python
import cv2
import dlib
from scipy.spatial import distance as dist
import numpy as np
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input
import pygame
import threading
import time
import datetime

# Constants for detection thresholds
EAR_THRESHOLD = 0.21  # Fine-tuned for sensitive detection
MAR_THRESHOLD = 0.6
YAWN_MIN_DURATION = 1.5  # Minimum duration for a yawn in seconds
MOUTH_STABILITY_THRESHOLD = 0.15  # Stability factor for consistent mouth opening
EYE_CLOSURE_DURATION = 7  # Duration (in seconds) for eye closure alert
EAR_CONSEC_FRAMES = 15  # Reduced for faster alerts
MAR_CONSEC_FRAMES = 10
ALERT_COOLDOWN_DURATION = 10  # Cooldown in seconds
log_file = "drowsiness_log.txt"

# Initialize counters and state flags
frame_counter_ear = 0
frame_counter_mar = 0
alert_active = False
alert_cooldown = False
last_alert_time = 0
yawn_start_time = None
previous_mar = 0.0
eye_closure_start_time = None

# Load Dlib's face detector and landmark predictor
detector = dlib.get_frontal_face_detector()
shape_predictor_path = "shape_predictor_68_face_landmarks.dat"
predictor = dlib.shape_predictor(shape_predictor_path)

# Build or load CNN model
def build_cnn_model():
    model = Sequential([
        Input(shape=(128, 128, 1)),  # Increased input size
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

try:
    cnn_model = build_cnn_model()
    print("CNN model loaded successfully.")
except Exception as e:
    print("Error loading CNN model: ", e)
    cnn_model = None

# Landmark indices
LEFT_EYE = list(range(36, 42))
RIGHT_EYE = list(range(42, 48))
MOUTH = list(range(60, 68))
ALL_LANDMARKS = list(range(68))

# Helper functions

def calculate_ear(eye):
    A = dist.euclidean(eye[1], eye[5])
    B = dist.euclidean(eye[2], eye[4])
    C = dist.euclidean(eye[0], eye[3])
    return (A + B) / (2.0 * C)


def calculate_mar(mouth):
    A = dist.euclidean(mouth[3], mouth[7])
    B = dist.euclidean(mouth[2], mouth[6])
    C = dist.euclidean(mouth[0], mouth[4])
    return (A + B) / (2.0 * C)


def extract_landmarks(shape, indices):
    return np.array([(shape.part(i).x, shape.part(i).y) for i in indices])


def adjust_brightness(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    value = np.mean(hsv[:, :, 2])
    correction_factor = 130 / (value + 1e-5)
    hsv[:, :, 2] = np.clip(hsv[:, :, 2] * correction_factor, 0, 255)
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)


def play_alert_sound():
    try:
        pygame.mixer.init()
        pygame.mixer.music.load("alert.mp3")
        pygame.mixer.music.play(-1)  # Loop indefinitely until stopped
    except pygame.error:
        print("Error: Unable to play alert sound.")


def stop_alert_sound():
    if pygame.mixer.get_init():
        pygame.mixer.music.stop()


def log_event(message):
    with open(log_file, "a") as file:
        file.write(f"[{datetime.datetime.now()}] {message}\n")


def display_alert(frame, message, position):
    cv2.putText(frame, message, position, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# Video capture setup
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Unable to access the webcam.")
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Adjust brightness for consistent detection
    frame = adjust_brightness(frame)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)

    # Reset cooldown if time elapsed
    if alert_cooldown and time.time() - last_alert_time > ALERT_COOLDOWN_DURATION:
        alert_cooldown = False
        print("Alert cooldown reset.")

    for face in faces:
        shape = predictor(gray, face)
        all_landmarks = extract_landmarks(shape, ALL_LANDMARKS)

        # Visual feedback on all detected landmarks
        for (x, y) in all_landmarks:
            cv2.circle(frame, (x, y), 2, (0, 255, 0), -1)  # Highlight all landmarks

        left_eye = extract_landmarks(shape, LEFT_EYE)
        right_eye = extract_landmarks(shape, RIGHT_EYE)
        mouth = extract_landmarks(shape, MOUTH)

        ear = (calculate_ear(left_eye) + calculate_ear(right_eye)) / 2.0
        mar = calculate_mar(mouth)

        # Check for sustained eye closure duration
        if ear < EAR_THRESHOLD:
            if eye_closure_start_time is None:
                eye_closure_start_time = time.time()
            elif time.time() - eye_closure_start_time >= EYE_CLOSURE_DURATION:
                display_alert(frame, "DROWSY! Eyes closed for too long!", (10, 30))
                log_event("Drowsiness detected (Eyes)")
                if not alert_active:
                    threading.Thread(target=play_alert_sound).start()
                    alert_active = True
        else:
            eye_closure_start_time = None

        # Check for yawning based on MAR, stability, and duration
        mar_difference = abs(mar - previous_mar)
        previous_mar = mar

        if mar > MAR_THRESHOLD and mar_difference < MOUTH_STABILITY_THRESHOLD:
            if yawn_start_time is None:
                yawn_start_time = time.time()
            elif time.time() - yawn_start_time >= YAWN_MIN_DURATION:
                display_alert(frame, "YAWNING ALERT! Take a break!", (10, 70))
                log_event("Drowsiness detected (Yawning)")
                if not alert_active:
                    threading.Thread(target=play_alert_sound).start()
                    alert_active = True
        else:
            yawn_start_time = None

        # Stop alert if no drowsiness detected
        if eye_closure_start_time is None and yawn_start_time is None and alert_active:
            stop_alert_sound()
            alert_active = False

    # Display status on the video frame
    cv2.putText(frame, "Drowsiness Detection Active", (10, 150),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

    # Display the frame
    cv2.imshow("Driver Drowsiness Detection", frame)

    # Exit on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()

stop_alert_sound()
print("Detection process complete. Check logs for drowsiness events.")
```
